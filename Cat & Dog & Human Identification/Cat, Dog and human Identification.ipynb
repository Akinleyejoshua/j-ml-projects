```python
import pickle as pk
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = pk.load(open("Dog cat human detection.pkl", "rb"))
```

```python
### load the image datasets, convert to numpy array, resize them to 1-dimentional array of (64,)
cat_img = np.array(plt.imread("cat.jpg"))
dog_img = np.array(plt.imread("dog.jpg"))
human_img = np.array(plt.imread("human.jpg"))
```

```python
# display the images and predict
plt.imshow(cat_img)
cat_img.resize(64, )
print("Prediction: ",model.predict([cat_img]))
```

```python
plt.imshow(dog_img)
dog_img.resize(64, )
print("Prediction: ",model.predict([dog_img]))
```

```python
plt.imshow(human_img)
human_img.resize(64, )
print("Prediction: ",model.predict([human_img]))
```
