### This classification program was built using Scikit-Learn module

```python
import numpy as np
import matplotlib.pyplot as plt
import pickle as pk
```

```python
# import the model
model = pk.load(open("Face mask model.pkl", "rb"))
```

```python
# import imgages and make predictions
no_face_masked_img = np.array(plt.imread("test1.jpg"))
plt.imshow(no_face_masked_img)
no_face_masked_img.resize(64, )
print(model.predict([no_face_masked_img]))
```

```python
face_masked_img = np.array(plt.imread("test2.jpg"))
plt.imshow(face_masked_img)
face_masked_img.resize(64, )
print(model.predict([face_masked_img]))
```
