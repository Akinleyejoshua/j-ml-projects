```python
# we are going to first import the libraries for this projects

# pandas is a data manipulation library which can be used to create, manipulate and view DataFrame from datasets
import pandas as pd

# MLPClassifier is a neural network model which would be used to fit in the datasets
from sklearn.neural_network import MLPClassifier

# for converting our data into a train and test datasets, train_test_split module must be imported from the sklearn library
from sklearn.model_selection import train_test_split

# import pandas for numeric computing
import numpy as np
```

```python
# Import the Heart Disease Datasets and feed it to Pandas DataFrame
heart_disease_df = pd.read_csv("heart-disease.csv")

# List the datasets features and target
heart_disease_df
```

### The feature sets are the sets which are independent sets which would be used for prediction
### The target set depends on the feature sets
### The feature sets which in this project would be identified as `X`
### The target sets would be identified as `y`

the feature sets include the `age`, `sex`, `cp` ...
the target sets include `0` and `1`
Since this is a Classification Project the target sets includes the
* `0` means Negative i.e Patient does not have Heart Disease
* `1` means Positive i.e Patient does have Heart Diseases

```python
# describe the dataset using describe function
heart_disease_df.describe()
```

## as we can see that the count is 303 which means the number of samples is `303`

```python
# store the feature sets into the variable X
# drop function removes an element from a DataFrame
# We do so by removing the target column(axis = 1)
X = heart_disease_df.drop("target", axis=1)
X.head()
```

```python
# store the target set into variable y
# we do so by selecting the target column from the dataframe
y = heart_disease_df["target"]
y.head()
```

```python
np.unique(heart_disease_df["target"])
# 0 : Negative
# 1: Positive
```

#### Spliting our datasets into train and test is very important 
* our train datasets takes about 80% while test sets about 20%

#### A train set is used to fit the data into the model
* It also used to allow the model to find pattern in a dataset

#### A test set is used to test our accuracy of our datasets, the better the accuracy the better the learned pattern
* A test set is used to test if the model can generalise on new data which it asn't seen
* The better accuracy the better the model can generalise on new data
* A test set can also serve as a validator to the model

```python
# so we are gonna create a train and test split for X and y
# setting test size to 20% === 0.2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

### Fitting the `X_train` and `y_train` to the `MLPClassifer` model
* this would make the model learn from the trained split data 
* Pattern are learned from the model

```python
# instantiate the model
model = MLPClassifier()

# fit the data into the model
model.fit(X_train, y_train)

# score the model to get it accuracy
accuracy = model.score(X_test, y_test)
print("Accuracy: ", accuracy)
```

we have an accuracy of about 85% which is also equalls 0.85, which means our model learned very well from the fitted data 
that also means the data can generalize from new data and predict well if a new patient has a heart disease

### Now we have our trained data we can now predict on our test data to check how well the model is doing
1. We would get the `X_test` and make predictions on the data
2. We would compare the `X_test` predictions with `y_test` to check if it actually predicted it accuratelly

```python
# make predictions on X_test
y_pred = model.predict(X_test)
print(y_pred)
```

```python
# compare predictions
boolean_pred = y_pred == y_test
boolean_pred
```

### As we can see from the predictions that some where predicted `True` and `False`
now let get the amount of predicted `false` and `true` using the `value_counts()` function

```python
boolean_pred.value_counts()
```

wow!!! our model perfomed well, we have more predictions to be accurate(`True`) than `False`
* `52` predictions where `True`
* `9` predictions where `False`

## Now let make predictions on new data the model have'nt seen on the feature sets given from the datasets below:
* `age` `sex` `cp` `trestbps` `chol`  `fbs` `restecg` `thalach` `exang` `oldpeak` `slope` `ca` `thal`
#### assuming the above data shows the patient health status that we want to make prediction on wether the patient have heart disease or not
the code: `name = model.predict([[age,sex,cp,trestbps,chol,fbs,restecg,thalach,exang,oldpeak,slope,ca,thal]])`
### Note: anywhere you see `1` it means `True` and `0` means `False` e.g `sex for 1 means Male which is True`

```python
# now to make it easily understood lets define a variable(the name of the patient) that stores the patient health data
dave = model.predict([[15, 1, 3, 120, 225, 1, 0, 155, 0, 1.5, 0, 0, 3]])
print("Prediction", dave[0])
```

Now let create a function that accepts the user data and displays if he or she has heart disease or his positive or negative

Since our prediction from the patient dave is `1 which means True` then `Dave` test result is `Positive` he has heart disease

```python
classes = ["Heart Diease: Negative", "Heart Disease: Positive"]
def get_pred_label(name, pred):
    pred = model.predict([pred])
    return f"{name} : {classes[pred[0]]}"
```

```python
dave = get_pred_label("Dave", [15, 1, 3, 120, 225, 1, 0, 155, 0, 1.5, 0, 0, 3])
mike = get_pred_label("Mike", [56, 1, 2, 100, 10, 1, 1, 122, 10, 4, 1, 1, 1])
print(dave)
print(mike)
```

the label result show that dave has heart disease while mike does'nt have heart disease
## Now let store our result in a DataFrame

```python
patients = ["Dave", "Mike"]
results = ["Positive", "Negative"]
patient_data = pd.DataFrame()
patient_data["Patients"] = pd.Series(patients)
patient_data["result"] = pd.Series(results)
patient_data
```
